```python
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
import pandas as pd
```

```python
df = sns.load_dataset('iris') # for MultiClass
#df =pd.read_csv('Data/placement.csv') for BinaryClass
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb_enode = LabelEncoder()
```

```python
df['species'] = lb_enode.fit_transform(df['species'])
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lg = LogisticRegression()
decT = DecisionTreeClassifier()
lg.fit(X_train,y_train)
decT.fit(X_train,y_train)
```

```python
ypred1 = lg.predict(X_test)
ypred2 = decT.predict(X_test)
```

```python
print("Logistic Accuracy = ", accuracy_score(y_test,ypred1))
print("DecisionTree Accuracy = ", accuracy_score(y_test,ypred2))
```
