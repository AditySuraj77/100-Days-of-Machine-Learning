```python
from sklearn.datasets import load_diabetes
from sklearn.datasets import make_regression
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data =  load_diabetes()
```

```python
import pandas as pd
```

```python
X,y = make_regression(n_samples=100,n_features=1,n_targets=1,n_informative=1,noise=20,random_state=13)
```

```python
#X= data.data
#y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=45)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
ypred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,ypred)
```

```python
mean_squared_error(y_test,ypred)
```

```python
plt.plot(X,y,'b. ')
plt.plot(X_train,lr.predict(X_train),color='red')
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
rid = Ridge(alpha=1.0)
```

```python
rid.fit(X_train,y_train)
```

```python
yridge_pred = rid.predict(X_test)
```

```python
r2_score(y_test,yridge_pred)
```

```python
mean_squared_error(y_test,yridge_pred)
```

```python
secondridge = Ridge(alpha=5)
```

```python
secondridge.fit(X_train,y_train)
```

```python
plt.plot(X,y,'b. ')
plt.plot(X_train,rid.predict(X_train),color='green', label='alpha=1.0')
plt.plot(X_train,lr.predict(X_train),color='red',label='Linear regression')
plt.plot(X_train,secondridge.predict(X_train),color='yellow',label='alpha=10')
```
