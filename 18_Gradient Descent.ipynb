```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X_test)
```

```python
print("MSE", mean_squared_error(y_test,y_pred))
print("r2", r2_score(y_test,y_pred))
```

# Build Own Gradient Descent Algos

```python
class MyGDRegressor:
    def __init__(self,learning_rate,epochs):
        self.m = 100
        self.b = -150
        self.learning_rate = learning_rate
        self.epochs = epochs


    def fit(self,X,y):
        for i in range(self.epochs):
            lose_slope_b = -2* np.sum(y - self.m*X.ravel() - self.b)
            lose_slope_m = -2* np.sum((y - self.m*X.ravel() - self.b)*X.ravel())
            self.b =  self.b - (self.learning_rate * lose_slope_b)
            self.m =  self.m - (self.learning_rate * lose_slope_m)
            #print(lose_slope,self.b)
            print("Intercept_ = ",self.b,"Coef_ = ", self.m)
            
```

```python
GDreg = MyGDRegressor(learning_rate=0.001,epochs=50)
```
